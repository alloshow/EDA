Напишем функцию, которая заменит номера дней недели на их названия
В функции создадим словарь, используя литерал {}, где обозначим ключи (нумерация) и значения (день недели).

```python
def change_day(row):
    list_days = {0: 'Понедельник', 1: 'Вторник', 2: 'Среда', 3: 'Четверг', 4: 'Пятница', 5: 'Суббота', 6: 'Воскресенье'}
    days = row['public_day']
    if days == 0:        
        days = list_days[0]  
        return days          
    elif days == 1:        
        days = list_days[1]
        return days        
    elif days == 2:
        days = list_days[2]
        return days      
    elif days == 3:
        days = list_days[3]
        return days      
    elif days == 4:
        days = list_days[4]
        return days
    elif days == 5:
        days = list_days[5]
        return days  
    elif days == 6:
        days = list_days[6]
        return days
    else:
        return days

data['public_day'] = data.apply(change_day, axis=1)
```

#### Добавим в таблицу столбец, в котором будет отображен тип этажа квартиры ("первый", "последний" или "другой")

```python
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
    
data['type_floor'] = data.apply(floor_category, axis = 1)
```

Напишем функцию, которая создаст столбец с категорией цены.

```python
def category_price(row):
    price = row['last_price']
    if price <= 3150.0:
        return 'очень дешевые'
    elif 3150.0 < price <= 3999.0:
        return 'дешевые'
    elif 3999.0 < price <= 4950.0:
        return 'средние'
    elif 4950.0 < price <= 6650.0:
        return 'дорогие'    
    elif price > 6650.0:
        return 'очень дорогие'  

data['category_price'] = data.apply(category_price, axis = 1)
```

```python
data_spb = data.query('locality_name == "Санкт-Петербург" and cityCenters_nearest_km ==3')
data_spb.pivot_table(index = 'category_price', values = 'cityCenters_nearest_km', aggfunc = 'count')
```
